```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop(columns=['id', 'Unnamed: 32'])
```

```python
df['diagnosis'] = df['diagnosis'].replace(['M', 'B'], [1, 0])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
correlations = df.corr(numeric_only=True)['diagnosis'].sort_values(ascending=False)
plt.figure(figsize=(8, 12))
sns.barplot(x=correlations.values, y=correlations.index)
plt.title('Correlation of Diagnosis with Other Features')
plt.show()
```

```python
df['diagnosis'].value_counts().plot(kind= 'bar')
```

```python
X = df_scaled.drop(['diagnosis'], axis=1)

y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
preditions = model.predict(x_test)
```

```python
preditions
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, preditions)
```

```python
print(f"The accuracy is : {accuracy : .2f}")
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, preditions)
print(report)
```
